```python
frames.index = pd.to_timedelta(frames.index, unit="ms")
frames.index.name = "time"
frames
```

```python
 # coordinates for left elbow:
frames[["left_wrist_y", "right_wrist_y"]]
```

```python
# select frames between 3rd and 4th second
start = pd.to_timedelta(3, unit="s")
stop = pd.to_timedelta(4, unit="s")
frames.loc[(frames.index >= start) & (frames.index < stop)]
```

```python
ELAN_ANNOTATION_FILE_PATH = "data/elan_data/text_files/arun_swipe_up_down.txt"
# read the ELAN file as CSV; only read the relevant columns (specified by `usecols`)
annotations = pd.read_csv(ELAN_ANNOTATION_FILE_PATH, sep="\t", header=None, usecols=[3, 5, 8], names=["start", "end", "label"])

# convert start and end timestamps into a datetime-datatype – this is important for the next step
annotations["start"] = pd.to_timedelta(annotations["start"], unit="s")
annotations["end"]   = pd.to_timedelta(annotations["end"], unit="s")

annotations
```

```python
# the default label will be 'idle'
frames["ground_truth"] = "idle"

# iterate over each entry from the ELAN file, select the corresponding frames and label them accordingly:
for idx, ann in annotations.iterrows():
    annotated_frames = (frames.index >= ann["start"]) & (frames.index <= ann["end"])
    frames.loc[annotated_frames, "ground_truth"] = ann["label"]

frames[["left_wrist_y", "right_wrist_y", "ground_truth"]]
```

```python
from pathlib import Path

# Create output directory if it doesn't exist
output_dir = Path("data/csv_results")
output_dir.mkdir(parents=True, exist_ok=True)

save_frames = frames.copy()

# convert index back to integers so it looks nicer in CSV
save_frames.index = save_frames.index.astype("int64") // 1_000_000

# save all data with the new ground_truth column
save_frames.to_csv(
    output_dir / "arun_swipe_up_poses.csv",
    index=True
)

# save only the ground_truth column
save_frames["ground_truth"].to_csv(
    output_dir / "arun_swipe_up_ground_truth.csv",
    index=True
)
```
